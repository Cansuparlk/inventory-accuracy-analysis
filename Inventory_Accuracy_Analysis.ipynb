```python
# STEP 1: Install required libraries
import pandas as pd
import random
import numpy as np

# STEP 2: Let's create product, warehouse and category lists
products = [f"P{str(i).zfill(3)}" for i in range(1, 101)]  # P001, P002, ...
warehouses = ["WH1", "WH2", "WH3", "WH4"]
categories = ["FMCG", "Electronics", "Beverages", "Apparel"]

# STEP 3: Empty data list
data = []

# STEP 4: Let's create 500 rows of sample data
for i in range(500):
    product = random.choice(products)
    warehouse = random.choice(warehouses)
    category = random.choice(categories)
    system_qty = random.randint(50, 300)
    deviation = random.randint(-30, 30)
    counted_qty = max(system_qty + deviation, 0)  # 0'dan küçük olmasın
    count_date = pd.Timestamp("2025-04-01") + pd.to_timedelta(random.randint(0, 30), unit='d')

    data.append([product, warehouse, system_qty, counted_qty, category, count_date])

# STEP 5: Convert to Pandas DataFrame
df = pd.DataFrame(data, columns=["product_id", "warehouse_id", "system_qty", "counted_qty", "category", "count_date"])

# STEP 6: Let's display the first 5 lines
df.head()
```

```python
df.to_csv("inventory_data.csv", index=False)
```
